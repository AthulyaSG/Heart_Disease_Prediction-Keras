```python
#Remove outliers
import matplotlib.pyplot as plt
import pandas as pd

column = ['resting blood pressure', 'cholesterol', 'maximum heart rate', 'ST depression']

# Function to cap outliers in a column using IQR
def cap_outliers_iqr(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower_cap = q1 - 1.5 * iqr
    upper_cap = q3 + 1.5 * iqr
    return column.clip(lower=lower_cap, upper=upper_cap)

# Select only the relevant columns for the box plot
selected_columns = df[column]

# Cap outliers using the IQR method
selected_columns = selected_columns.apply(cap_outliers_iqr)

# Create subplots with 2 rows
num_rows = 2
num_cols = (len(column) + num_rows - 1) // num_rows

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))
fig.suptitle('Box Plots After Capping Outliers', y=1.02)

# Flatten the 2D array of subplots into a 1D array for easy indexing
axes = axes.flatten()

# Box plots for each numerical column
for i, column in enumerate(selected_columns.columns):
    selected_columns[column].plot(kind='box', ax=axes[i])
    axes[i].set_title(f'Box Plot for {column}')
    axes[i].set_ylabel('Values')

# Remove empty subplots
for i in range(len(column), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from keras.models import Sequential
from keras.layers import Dense

# Selecting features
X = df_filled.drop(['heart disease'], axis = 1)
y = df_filled['heart disease']

# Getting column names
col_names = list(X.columns)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Creating a DataFrame with scaled features
X_scaled = pd.DataFrame(X_scaled, columns=col_names)

# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=40)

```

```python
# Balance the training for generalization without overfitting
from keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(
    min_delta=0.001,  
    patience=20,     
    restore_best_weights=True 
)

# Build Neural Network Model
model = Sequential()

# Add the input layer with 5 neurons
model.add(Dense(units=5, kernel_initializer='uniform', activation='relu', input_dim=13))

# Add the first hidden layer with 5 neurons
model.add(Dense(units=5, kernel_initializer='uniform', activation='relu'))
model.add(Dropout(0.25))

# Add the second hidden layer with 5 neurons
model.add(Dense(units=5, kernel_initializer='uniform', activation='relu'))
model.add(Dropout(0.25))

# Add the third hidden layer with 3 neurons
model.add(Dense(units=3, kernel_initializer='uniform', activation='relu'))
model.add(Dropout(0.25))

# Output layer with 1 neuron (binary outcome) and sigmoid activation
model.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the Model
model.fit(X_scaled, y, epochs=30, batch_size=5, callbacks=[early_stopping], validation_split=0.2)
```
